<h5> Q2.In which year were the highest number of TV shows and movies released? Show the results with the help of visualization (Bar Graph) </h5>

```python
data.dtypes
```

to_datetime()

```python
data['Date_N'] = pd.to_datetime(data['Release_Date']) # to convert 'Release_Date' column to datetime and create new column 'Date_N'
```

```python
data.head()
```

```python
data.dtypes
```

data.year.value_counts()

```python
data['Date_N'].dt.year.value_counts()
```

Bar Graph

```python
plt.figure(figsize=(10,8))
sns.barplot(x = data['Date_N'].dt.year.value_counts().index, 
            y = data['Date_N'].dt.year.value_counts(), 
            order= data['Date_N'].dt.year.value_counts().sort_values(ascending = False).index,
            palette = 'rainbow')
plt.xlabel('Date_N', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.title('Number of movies released each year', fontsize = 20)
plt.show()
```

The year of 2019 witnessed the maximum number of TV Shows and Movies release with 2154 TV Shows and Movies released in that year.

<h5> Q3. How many movies and TV shows are in the dataset? Visualize with the help of a bar graph.</h5>

```python
data.groupby('Category')['Title'].count() # to show number of movies and TV Shows released each year
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x = data['Category'].value_counts().index,
            y = data['Category'].value_counts().values,
            palette = 'viridis', order= data['Category'].value_counts().sort_values(ascending = True).index)
plt.xlabel('Category', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.title('Number of movies and TV Shows released each year', fontsize = 20)
plt.show()
```

Using countplot

```python
plt.figure(figsize=(10,8))
sns.countplot(x = data['Category'], palette='rainbow')
plt.xlabel('Category', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.title('Number of movies and TV Shows released each year', fontsize = 20)
plt.show()
```

As can be seen from the bar graph, there are 5378 movies and 2410 TV shows available in the dataset.

<h5> Q6. Show the top 10 Directors who gave the highest number of TV Shows and Movies to Netflix. </h5>

```python
data.groupby('Director')['Title'].count().sort_values(ascending = False).head(10)
```

Visualizing the results using countplot

```python
plt.figure(figsize=(17,16))
sns.countplot(x = data['Director'], order = data['Director'].value_counts().index[:10], palette='rainbow')
plt.xlabel('Director', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.title('Top 10 directors with most number of movies', fontsize = 20)
plt.show()
```

The director "Raúl Campos, Jan Suter" has given the highest number of TV Shows and Movies to Netflix.
